```python
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import requests


url="https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc"
req=requests.get(url)
soup=BeautifulSoup(req.content,'html.parser')


```

```python
movie_name=[]
rating=[]
year=[]
```

```python
movie_data=soup.find_all('div',attrs={'class':'lister-item mode-advanced'})
movie_data

                                      
```

```python
for store in movie_data:
    name=store.h3.a.text
    movie_name.append(name)    
    
    year_of_release=store.h3.find('span',class_='lister-item-year text-muted unbold').text.replace("(",'').replace(')','')                                                                                      
    year.append( year_of_release)

    rating_data=store.find('div',class_="inline-block ratings-imdb-rating").text.replace('\n','')
    rating.append(rating_data)
    
    
```

```python
movie_df=pd.DataFrame({'Name of movie':movie_name,'Year of release':year,'Rating':rating})
```
