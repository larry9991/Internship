```python
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import requests


url="https://www.journals.elsevier.com/artificial-intelligence/most-downloaded-articles"
req=requests.get(url)
soup=BeautifulSoup(req.content,'html.parser')
```

```python
papertitle=[]
authors=[]
publisheddate=[] 
paperurl=[]
```

```python
site_data=soup.find_all('li',attrs={"class":"sc-9zxyh7-1 sc-9zxyh7-2 exAXfr jQmQZp"})
site_data
```

```python
for store in site_data:
    title=store.h2.text
    papertitle.append(title)
    
    aut=store.p.find('span',class_="sc-1w3fpd7-0 pgLAT").text
    authors.append(aut)
    
    pub=store.find('span',class_="sc-1thf9ly-2 bKddwo").text
    publisheddate.append(pub)
    
```

```python
paper_df=pd.DataFrame({'PaperTitle':papertitle,'Authors':authors,'Published_date':publisheddate})
```
