```python
import numpy as np
import pandas as pd
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings ('ignore')
import time
```

```python
driver=webdriver.Chrome(r'D:\chromedriver.exe')
```

```python
url='https://trak.in/'
driver.get(url)
```

```python
search_bt=driver.find_element(By.XPATH,"//*[@id='main-navigation']/li[9]/a")
search_bt.click()
```

```python
search_btf=driver.find_element(By.XPATH,"//*[@id='menu-item-51510']/a")
search_btf.click()
```

```python
brand=[]
start=0
end=3
for page in range(start,end):
    brand_tags=driver.find_elements(By.CLASS_NAME,'row-hover')
    for i in brand_tags:
        brand.append(i.text)
  
```
